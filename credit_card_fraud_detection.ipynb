```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,r2_score
```

Read data

```python
credit_data=pd.read_csv("creditcard.cv")
credit_data
```

```python
numeric_data = credit_data.select_dtypes(include=['number'])
numeric_data.corr()
f, ax = plt.subplots()
sns.heatmap(numeric_data.corr(), annot=True, linewidth = .5, fmt='.1f', ax=ax)
```

Check for null values

```python
credit_data.isnull().sum()
```

No null values

```python
credit_data['Class'].value_counts()
```

Split into train and test data

```python
x=credit_data.drop(columns='Class')
x=x.drop(columns='Amount')
y=credit_data['Class']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

Create model

```python
model1=LogisticRegression()
model2=RandomForestClassifier()
```

```python
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
```

predict values

```python
log_pred=model1.predict(x_test)
ran_pred=model2.predict(x_test)
x_test
```

accuracy and r2 score

```python
print("Logistic Regression: ")
print("Accuracy score: ",accuracy_score(y_test,log_pred))
print("R2_score: ",r2_score(y_test,log_pred))
print("Random Forest Regressor: ")
print("Accuracy score: ",accuracy_score(y_test,ran_pred))
print("R2_score: ",r2_score(y_test,ran_pred))
```

Random Forest Classifier givess 99.95% accuracy so it is used for prediction

```python
model=model2
```
